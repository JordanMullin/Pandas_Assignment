```python
top_spenders_df = df.copy()
top_spenders_df.head()
```

```python
spenders_price_grouped_df = top_spenders_df.groupby("SN")["Price"].sum().reset_index()
```

```python
spenders_price_grouped_df.sort_values("Price", ascending=False)
```

```python
spenders_price_grouped_df.nlargest(5, "Price")
```

```python
spenders_count_grouped_df = top_spenders_df.groupby("SN")["Purchase ID"].count().reset_index()
spenders_count_grouped_df.sort_values("Purchase ID", ascending=False)
```

```python
#concatenating previous dataframes to obtain total purchases and total spent per player
concat_df = pd.concat([spenders_count_grouped_df["Purchase ID"], spenders_price_grouped_df], axis=1)
concat_df = concat_df[["SN", "Purchase ID", "Price"]]
concat_df.sort_values("Price", ascending=False)
```

```python
#finishing concatenation but still need to rename 'price' column
concat_df["Average Purchase Price"] = concat_df["Price"]/concat_df["Purchase ID"]
concat_df.sort_values("Price", ascending=False)
org_concat_df = concat_df[["SN", "Purchase ID", "Average Purchase Price", "Price"]]
org_concat_df
```

```python
#renaming and presenting the final purchase summary by top spenders
renamed_concat_df = org_concat_df.rename(columns = {"Purchase ID": "Purchase Count", "SN":"SN", "Average Purchase Price":"Average Purchase Price", "Price":"Total Purchase Value"})
renamed_concat_df.sort_values("Total Purchase Value", ascending=False).head()
```

```python
#grouping data together
pop_df = df.copy()
item_pop_df = pop_df.groupby("Item ID")["Item Name"].unique()
item_pop_df
```

```python
#grouping data together
count_pop_df = pop_df.groupby("Item ID").nunique()
ren_count_pop_df = count_pop_df.rename(columns={
    "Purchase ID":"Purchase Count"})
purch_count_pop_df = ren_count_pop_df[["Purchase Count"]]
purch_count_pop_df
```

```python
#grouping data together
price_pop_df = pop_df.groupby("Item ID")["Price"].unique()
price_pop_df
```

```python
#grouping data together
price_pop_id_df = pop_df.groupby("Item ID")["Price"].sum()

ren_price_pop_id_df = price_pop_id_df.rename(columns = {
    "Price":"Total Purchase Value"})
ren_price_pop_id_df
```

```python
#concatenating groups together
pop_concat_df = pd.concat([item_pop_df, purch_count_pop_df], axis=1)

pop_concat_df
```

```python
#concatenation cont.
pop_concat_df2 = pd.concat([pop_concat_df, price_pop_df], axis=1)
pop_concat_df2
```

```python
#final concatenation and renaming columns appropriately
pop_concat_df3 = pd.concat([pop_concat_df2, ren_price_pop_id_df], axis=1)
ren_pop_concat_df3 = pop_concat_df3.rename(columns = {
    0:"Total Purchase Value"})
ren_pop_concat_df3

```

```python
#presenting a summary of the most popular purchases sorted by purchase count
ren_popular_purchases_df = ren_pop_concat_df3.rename(columns={
    "Price":"Item Price"})
popular_purchases_df = ren_popular_purchases_df.nlargest(5, "Purchase Count")
popular_purchases_df
```
